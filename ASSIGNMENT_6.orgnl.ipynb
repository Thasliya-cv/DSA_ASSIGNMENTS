# REGRESSION

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('car_age_price.csv')
df
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
sns.lineplot(x='Year',y='Price',data=df)
```

```python
x=df['Year'].values.reshape(-1,1)
y=df["Price"].values.reshape(-1,1)
```

```python
x_train,y_train,x_test,y_test=train_test_split(x,y)
```

```python
lin=LinearRegression()
```

```python
lin_model=lin.fit(x,y)
lin_pred=lin_model.predict([[2022]])
lin_pred
```

```python
plt.scatter(x,y,color='blue',label='Datas')
plt.plot(x, lin.predict(x), color='red', label='Regression Line')
plt.scatter(2022,lin_pred,color='green',label='2022 Prediction')
plt.xlabel('Year')
plt.ylabel('Price')
plt.legend()
plt.show()
```
