```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('iris.csv')
df.head()
```

```python
df1=df.select_dtypes(float)
df1
```

### FUNCTION TRANSFORMERS

#### Log transform

```python
from sklearn.preprocessing import FunctionTransformer
log=FunctionTransformer(func=np.log1p)
log_transform=log.fit_transform(df1)
log_transform.hist()
log_transform.skew()
```

#### Square transform

```python
from sklearn.preprocessing import FunctionTransformer
sqr=FunctionTransformer(func=np.square)
sqr_transform=sqr.fit_transform(df1)
sqr_transform.hist()
sqr_transform.skew()
```

#### square root transform

```python
from sklearn.preprocessing import FunctionTransformer
sqrt=FunctionTransformer(func=np.sqrt)
sqrt_transform=sqrt.fit_transform(df1)
sqrt_transform.hist()
sqrt_transform.skew()
```

#### sine transform

```python
from sklearn.preprocessing import FunctionTransformer
sine=FunctionTransformer(func=np.sin)
sine_transform=sine.fit_transform(df1)
sine_transform.hist()
sine_transform.skew()
```

#### cosine transform

```python
from sklearn.preprocessing import FunctionTransformer
cosine=FunctionTransformer(func=np.cos)
cosine_transform=cosine.fit_transform(df1)
cosine_transform.hist()
cosine_transform.skew()
```

#### Tan transform

```python
from sklearn.preprocessing import FunctionTransformer
tan=FunctionTransformer(func=np.tan)
tan_transform=tan.fit_transform(df1)
tan_transform.hist()
tan_transform.skew()
```

### POWER TRANSFORMER

#### box cox transform

```python
from sklearn.preprocessing import PowerTransformer
boxcox=PowerTransformer(method='box-cox')
boxcox_transform=boxcox.fit_transform(df1)
boxcox_transform=pd.DataFrame(boxcox_transform)
boxcox_transform.hist()
boxcox_transform.skew()
```

#### yeo jhonson transform

```python
from sklearn.preprocessing import PowerTransformer
yeo=PowerTransformer()
yeo_transform=yeo.fit_transform(df1)
yeo_transform=pd.DataFrame(yeo_transform)
yeo_transform.hist()
yeo_transform.skew()
```

### QUANTILE TRANSFORM

```python
from sklearn.preprocessing import QuantileTransformer
Q=QuantileTransformer(output_distribution='normal')
Q_transform=Q.fit_transform(df1)
Q_transform=pd.DataFrame(Q_transform)
Q_transform.hist()
Q_transform.skew()
```

#### ***YEO jHONSON transformer is most suitable***
